```python
import pandas as pd

data= pd.DataFrame({ 'OrderID': [1, 2, 3, 4, 5],
'CustomerName': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'], 'Category':
['Electronics', 'Apparel', 'Electronics', 'Home', 'Apparel'], 'OrderAmount':
[500, 200, 1500, 300, 700], 'Discount': [50, 20, 100, 30, 70], 'OrderDate':
['2024-11-01', '2024-11-02', '2024-11-03', '2024-11-04', '2024-11-05'],
'Region': ['North', 'South', 'North', 'West', 'East'] })

def categorized_order(amount):
    if amount > 1000:
        return 'high'
    elif 500<= amount <=1000:
        return 'Medium'
    else:
        return 'Low'
df=  data.loc[:, 'OrderAmount'].apply(categorized_order)
print(df)
```
